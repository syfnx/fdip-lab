## 1. To detect horizontal and vertical lines in an image
## Line detections:
- Horizontal lines
- Vertical lines
- -45° lines
- 45° lines

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
folder = "/content/drive/MyDrive/Colab Notebooks/img/"
```

```python
path = folder + "lines.png"
flag = 0
img = cv2.imread(path, flag)
```

```python
cv2_imshow(img)
```

```python
kernel = np.ones((2,19), np.uint8)
hl = cv2.erode(img, kernel, iterations=1)
res = np.hstack((img,hl))
print("Horizontal lines: ")
cv2_imshow(res)
```

```python
kernel = np.ones((19,2), np.uint8)
vl = cv2.erode(img, kernel, iterations=1)
res = np.hstack((img,vl))
print("Vertical lines: ")
cv2_imshow(res)
```

```python
kernel = np.zeros((20,20), np.uint8)
for i in range(len(kernel)):
  kernel[i][i] = 1

l45 = cv2.erode(img, kernel, iterations=1)
res = np.hstack((img,l45))
print("-45° lines: ")
cv2_imshow(res)
```

```python
kernel = np.zeros((20,20), np.uint8)
n = len(kernel)
for i in range(n):
  kernel[n-i-1][i] = 1
l_45 = cv2.erode(img, kernel, iterations=1)
res = np.hstack((img,l_45))
print("45° lines: ")
cv2_imshow(res)
```

## 2. To illustrate Houghline method for line detection 

```python
img = cv2.imread(folder + "parking_lot.jpg", 0)
edge = cv2.Canny(img,100,200)
hl = cv2.HoughLines(edge, 1, np.pi/180, 200)

lines_list =[]
lines = cv2.HoughLinesP(edge, 1, np.pi/180, threshold=100, minLineLength=5, maxLineGap=10)
  
# Iterate over points
for points in lines:
    x1,y1,x2,y2 = points[0]
    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
    lines_list.append([(x1,y1),(x2,y2)])

print("HoughLines: ")
cv2_imshow(img)
```

## 3. To illustrate Canny edge detection algorithm

```python
img = cv2.imread(folder + "tiger.jpg", 0)
edge = cv2.Canny(img,100,200)

res = np.hstack((img,edge))
print("Canny: ")
cv2_imshow(res)
```
