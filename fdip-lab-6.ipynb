### AIM: To study and perform different types of Morphological Operations
1. Dilation
2. Erosion 
3. Opening
4. Closing

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
folder = "/content/drive/MyDrive/Colab Notebooks/img/"
```

```python
path = folder + "lenna.png"
flag = 0
img = cv2.imread(path, flag)

# Taking a matrix of size 5 as the kernel
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
print("kernel: \n",kernel)
print()

img_dilation = cv2.dilate(img, kernel, iterations=1)

res = np.hstack((img,img_dilation))
print("Dilation:")
cv2_imshow(res)
```

```python

img_erosion = cv2.erode(img, kernel, iterations=1)
res = np.hstack((img,img_erosion))

print('Erosion: ')
cv2_imshow(res)
```

```python
path = folder + "opening.jpg"
flag = 0
img = cv2.imread(path, flag)
```

```python
# opening
# erosion -> dilation

op = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=25)

res = np.hstack((img,op))
print("Opening: ")
cv2_imshow(res)
```

```python
# closing
# erosion -> dilation
op = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=21)

res = np.hstack((img,op))
print("Closing: ")
cv2_imshow(res)
```
